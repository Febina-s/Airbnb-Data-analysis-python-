## EDA

**Task 5. Data analysis**

**Univariate Analysis** 
#### we are analysing one column at a time and knowing its distribution (speding in dataset)

```python
#Price distribution

sns.histplot(data=data,x='price')
```

```python
sns.boxplot(data=data,x='price')
```

```python
#Identifying outliers in price

df=data[data['price']<1500]
sns.boxplot(data=df,x='price')
```

```python
#Price Distribution
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='price',color='maroon',bins=110)
plt.ylabel('Frequency')
```

```python
df.columns
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

Feature Engineering 

```python
#price per bed

df['price per bed']=df['price']/df['beds']
df.head()
```

```python
#now the difference can see 
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

***Bivariate Analysis***
  (one variable dependencies to other)

```python
#Price vs number_of_reviews
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='number_of_reviews',y='price',alpha=0.6, hue='neighbourhood_group')
plt.title('Locality and Review dependancy')
plt.show()
```

```python
#Avg price per neighbourhood groups 

plt.figure(figsize=(6,5))
sns.barplot(data=data,x='neighbourhood_group',y='price',estimator='mean')
plt.xticks(rotation=30)
plt.title('Average Price across Neighbourhood Groups')
plt.show()
```

```python
## Pairplot to show relationships and distributions among key numerical features

cols = ['price', 'minimum_nights', 'number_of_reviews', 
        'reviews_per_month', 'availability_365']

sns.pairplot(data[cols], diag_kind='kde')
plt.show()
```

```python
# Heatmap to visualize correlations between numerical features

numeric_cols = data.select_dtypes(include=['float64','int64'])
plt.figure(figsize=(8,8))
sns.heatmap(numeric_cols.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Countplot showing room type distribution across neighbourhood groups

plt.figure(figsize=(10,6))
sns.countplot(data=data, x='room_type', hue='neighbourhood_group')
plt.title("Room Type Distribution by Neighbourhood Group")
plt.show()
```

```python
# Boxplot to compare price distribution across different bedroom counts
data['bedrooms'] = pd.to_numeric(data['bedrooms'], errors='coerce')

plt.figure(figsize=(8,5))
sns.boxplot(data=data, x='bedrooms', y='price')
plt.title("Price vs Bedrooms")
plt.show()
```
